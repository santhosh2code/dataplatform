This is to generate User Behaviour data for different ads plans and to get performance of each ad. Fields in this dataset would be
user_id, page_views,	time_spent_seconds,	ads_clicked,	date,	device_type,	plan_type

```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime, timedelta
import random
```

```python
fake = Faker()

num_records = 50000
devices = ['Mobile', 'Desktop', 'Tablet']
plan_types = ['Basic', 'Standard', 'Premium', 'None']  # Possible plan types
```

```python
def random_date(start_date, end_date):
    time_between = end_date - start_date
    random_days = np.random.randint(0, time_between.days)
    return start_date + timedelta(days=random_days)
```

```python
data = {
    'user_id': [fake.uuid4() for _ in range(num_records)],  # Random unique user IDs
    'page_views': np.random.randint(1, 20, num_records),  # Random page views (1 to 20)
    'time_spent_seconds': np.random.randint(10, 600, num_records),  # Random time spent (10 to 600 seconds)
    'ads_clicked': np.random.randint(0, 5, num_records),  # Random ads clicked (0 to 5)
    'date': [random_date(datetime(2023, 1, 1), datetime(2023, 12, 31)) for _ in range(num_records)],  # Random date in 2023
    'device_type': np.random.choice(devices, num_records),  # Random device type
    'plan_type': np.random.choice(plan_types, num_records)  # Random plan type
}
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df.to_csv('user_behaviour_data.csv', index=False)
```
