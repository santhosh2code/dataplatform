```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime,timedelta
```

```python
fake = Faker()
```

```python
num_records = 10000
channels = ['Facebook', 'Instagram', 'Twitter', 'YouTube','Google','SMSAds','Display Ads','Email','Direct mail']
#campaign_names = [f"Campaign_{i}" for i in range(1,20)]
```

```python
def random_date(start_date, end_date):
    time_between = end_date - start_date
    random_days = np.random.randint(0, time_between.days)
    return start_date + timedelta(days=random_days)
```

```python
date = {
    'campaign_id' : np.random.randint(1,20,num_records),
    'campaign_name': [fake.catch_phrase() for _ in range(num_records)],
    'channel': np.random.choice(channels, num_records),
    'impressions': np.random.randint(100,10000,num_records),
    'clicks': np.random.randint(10,500,num_records),
    'conversions': np.random.randint(1,50,num_records),
    'revenue': np.random.uniform(10,500,num_records),
    'date': [random_date(datetime(2023, 1, 1), datetime(2023, 12, 31)) for _ in range(num_records)]
}
```

```python
df = pd.DataFrame(date)
```

```python
df.to_csv('Campaign_data.csv',index=False)
```
