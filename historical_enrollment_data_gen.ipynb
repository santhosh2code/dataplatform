```python
import pandas as pd
from faker import Faker
import numpy as np
```

```python
fake = Faker()
rec_count = 100000
```

```python
gender = ['Male', 'Female', 'Other']
inc_level = ['Low', 'Medium', 'High']
enroll_outcome = ['1','0']
past_camp_data = ['Email Opened', 'Clicked Ad', 'Attended Webinar']
```

```python
hist_ds = {
    'prospect_id': [fake.uuid4() for i in range(rec_count)],
    'age': [fake.random_int(min=18, max=65) for i in range(rec_count)],
    'gender': np.random.choice(gender, rec_count),
    'income_level': np.random.choice(inc_level, rec_count),
    'engagement_score': [fake.random_int(min=1, max=100) for i in range(rec_count)],
    'past_campaign_data': np.random.choice(past_camp_data, rec_count),
    'number_of_previous_enroll_attmpts' : [fake.random_int(min=0, max=5) for i in range(rec_count)],
    'enroll_outcome': np.random.choice(enroll_outcome, rec_count)
}
```

```python
df = pd.DataFrame(hist_ds)
```
