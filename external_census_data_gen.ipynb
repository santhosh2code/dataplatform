```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
fake = Faker()

# Parameters
num_records = 100000
```

```python
data = {
    'location_id': [fake.uuid4() for _ in range(num_records)],  # Unique location ID
    'city': [fake.city() for _ in range(num_records)],  # Random city names
    'state': [fake.state() for _ in range(num_records)],  # Random state names
    'population': np.random.randint(1000, 1000000, num_records),  # Random population (1000 to 1,000,000)
    'median_age': np.round(np.random.uniform(20, 50, num_records), 1),  # Random median age (20 to 50)
    'median_income': np.round(np.random.uniform(30000, 120000, num_records), 2),  # Random median income ($30,000 to $120,000)
    'unemployment_rate': np.round(np.random.uniform(2, 15, num_records), 2),  # Random unemployment rate (2% to 15%)
    'poverty_rate': np.round(np.random.uniform(5, 30, num_records), 2),  # Random poverty rate (5% to 30%)
    'life_expectancy': np.round(np.random.uniform(65, 85, num_records), 1),  # Random life expectancy (65 to 85 years)
    'obesity_rate': np.round(np.random.uniform(10, 40, num_records), 2),  # Random obesity rate (10% to 40%)
    'smoking_rate': np.round(np.random.uniform(5, 25, num_records), 2),  # Random smoking rate (5% to 25%)
    'healthcare_access_score': np.round(np.random.uniform(0, 100, num_records), 2)  # Random healthcare access score (0 to 100)
}
```

```python
df = pd.DataFrame(data)
```
